```
import nltk
```

```
sent = "the quick brown fox jumps over the lazy dog the quick brown cat ran away the quick car drive fast"
words = sent.split(" ")
```

```
print(words)
```

```
word_bigrams = nltk.bigrams(words)
```

```
condition_pairs = [ ( (bigram[0],), bigram[1] ) for bigram in word_bigrams ] # ((w1),w2)
```

```
print(condition_pairs) # p(w2|w1)
```

```
cfd = nltk.ConditionalFreqDist(condition_pairs)
```

```
cfd.conditions() # all the w1 in p(w2|w1)
```

```
# syntax is cfd[condition][next word]
cfd[('the',)]['quick'] # count the occurance of "the quick"
```

```
cfd[('the',)]
```

```
cfd[('the',)].values()
```

```
sum(cfd[('the',)].values()) # count the occurance of "the quick". We have "the quick brown fox", "the quick brown cat" and "the quick car"
```

```
# w1 = the, w2 = quick, w3 = brown
# p(w3|w2,w1) = p(brown|quick,the)
prob = cfd[('the',)]['quick'] / sum(cfd[('the',)].values())
print(prob)
```
