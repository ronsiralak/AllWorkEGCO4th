# Calculate $\theta_0$ and $\theta_1$

```python
xbar = np.mean(X)
ybar = np.mean(Y)
```

```python
Theta1 = np.sum((X - xbar) * (Y - ybar)) / np.sum(np.power(X - xbar,2))
Theta0 = ybar - Theta1 * xbar
```

$f(x) = \theta_0 + \theta_1x$

# Building Model from Scratch

```python
class MyModel(tf.Module):
  def __init__(self, **kwargs):
    super().__init__(**kwargs)
    self.theta0 = tf.Variable(np.random.uniform(-5,5))
    self.theta1 = tf.Variable(np.random.uniform(-5,5))
    self.theta2 = tf.Variable(np.random.uniform(-5,5))

  def __call__(self, x):
    return self.theta2*tf.math.sin(self.theta1*x)  + self.theta0
    #return self.theta1 * x + self.theta0
```

```python
def loss(target, predicted):
  return tf.reduce_mean(tf.square(target - predicted))
```

$\theta_j = \theta_j - \alpha \frac{\partial J}{\partial \theta_j}$

```python
def train(model, x, y, learning_rate = 0.1):
  with tf.GradientTape() as tape:
    current_loss = loss(y, model(x))
  
  d_theta0, d_theta1, d_theta2 = tape.gradient(current_loss, [model.theta0, model.theta1,model.theta2])

  model.theta0.assign_sub(learning_rate * d_theta0)
  model.theta1.assign_sub(learning_rate * d_theta1)
  model.theta2.assign_sub(learning_rate * d_theta2)
```

```python
def training_loop(model, x, y, epochs, alpha):
  for epoch in range(1, epochs+1):
    if epoch % 100 == 0:
      alpha = alpha * 0.9 
    train(model, x, y, learning_rate = alpha)
```

```python
model = MyModel()
training_loop(model, X, Y , 1000, 0.3)
```
