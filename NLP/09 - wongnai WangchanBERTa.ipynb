# Main

```python
import numpy as np
import re
import pandas as pd
import torch
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_fscore_support, accuracy_score
from transformers import RobertaForSequenceClassification, Trainer, TrainingArguments, AutoTokenizer, CamembertTokenizer
from transformers import EarlyStoppingCallback
```

```python
class IntoDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)


def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='weighted')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }
```

```python
data_path = "/content/w_review_train.csv"
```

```python
with open(data_path) as f:
    lines = f.readlines()

reviews_long_str = " ".join(lines)
reviews_long_lst = reviews_long_str.split(";")

texts = []
stars = []

for i in np.arange(0, len(reviews_long_lst)-1):
    if len(reviews_long_lst[i+1]) >= 1: # exclude empty reviews
        if reviews_long_lst[i+1][0] in set(['1','2','3','4','5']): # some reviews don't have score, check first if score is given
            text = reviews_long_lst[i]

            # clean text
            if i > 0:
                text = text[1:]
            text = re.sub("\"", "", text)
            text = re.sub("\n", " ", text)
            text = text.strip()

            # store for make dataframe
            texts.append(text)
            stars.append(int(reviews_long_lst[i+1][0]))
```

```python
df = pd.DataFrame({
    "review":texts,
    "star":stars
})
```

## Tokenize

```python
labelencoder = LabelEncoder()
lb = labelencoder.fit(stars)
y = lb.transform(stars)
x_train, x_test, y_train, y_test = train_test_split(texts, y, test_size=0.2, random_state=12345, shuffle=True)
```

```python
tokenizer = AutoTokenizer.from_pretrained("airesearch/wangchanberta-base-att-spm-uncased", model_max_length=505)

train_encodings = tokenizer(x_train, truncation=True, padding=True)
test_encodings = tokenizer(x_test, truncation=True, padding=True)

train_dataset = IntoDataset(train_encodings, y_train)
val_dataset = IntoDataset(test_encodings, y_test)
```

## Load pretrained model

```python
model = RobertaForSequenceClassification.from_pretrained('airesearch/wangchanberta-base-att-spm-uncased', num_labels=5)
model.train()
```

## Train configs

```python
cb = EarlyStoppingCallback(early_stopping_patience=3)

training_args = TrainingArguments(
    output_dir="/content/wongnai_checkpoints",
    num_train_epochs=10,
    per_device_train_batch_size=8,
    do_eval=True,
    save_steps=200,
    save_total_limit=3,
    evaluation_strategy="steps",
    load_best_model_at_end=True,
    metric_for_best_model='f1',
    greater_is_better=True,
    weight_decay=5e-05,
    eval_steps=200,
    learning_rate=2e-05,
    seed=12345,
    lr_scheduler_type="linear",
    warmup_steps=200,
    logging_dir="/content/wongnai_log"
)
```

```python
trainer = Trainer(
    model=model,                         
    args=training_args,                  
    train_dataset=train_dataset,         
    eval_dataset=val_dataset,            
    compute_metrics=compute_metrics,
    callbacks=[cb]
)
```

# train and eval

```python
trainer.train()
print(trainer.evaluate())
trainer.save_model("/content/wongnai_save")
```
