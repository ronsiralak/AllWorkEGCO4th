# Thai Tokenization

We show 3 libraries for Thai tokenization (ตัดคำภาษาไทย)

deepcut https://github.com/rkcosmos/deepcut (tf 2.x)

cutkum https://github.com/pucktada/cutkum (tf 1.x)

pythainlp https://github.com/PyThaiNLP/pythainlp (pytorch)

```
import deepcut
toks = deepcut.tokenize('ตัดคำได้ดีมาก')
print(toks)
```

```
text = 'ตารางธาตุ (ในภาพ) คือ การจัดเรียงธาตุเคมีในรูปแบบของตารางตามเลขอะตอม การจัดเรียงอิเล็กตรอน และสมบัติทางเคมีที่ซ้ำกัน โดยจะใช้แนวโน้มพิริออดิกเป็นโครงสร้างพื้นฐานของตาราง แถวแนวนอนทั้ง 7 ของตารางเรียกว่า "คาบ" โดยปกติโลหะอยู่ฝั่งซ้ายและอโลหะอยู่ฝั่งขวา ส่วนแถวแนวตั้งเรียกว่า "หมู่" ประกอบด้วยธาตุที่มีสมบัติทางเคมีคล้ายคลึงกัน มี 6 หมู่ที่ได้รับการตั้งชื่อที่ยอมรับกันทั่วไปและเลขหมู่ เช่น ธาตุหมู่ 17 มีชื่อว่า แฮโลเจน และธาตุหมู่ 18 มีชื่อว่า แก๊สมีตระกูล ตารางธาตุยังมีอาณาเขตรูปสี่เหลี่ยมผืนผ้าอย่างง่าย 4 รูปที่เรียกว่า "บล็อก" ซึ่งเกี่ยวข้องกับการเติมออร์บิทัลเชิงอะตอมที่แตกต่างกัน ธาตุทุกตัวนับตั้งแต่มีเลขอะตอมเท่ากับ 1 (ไฮโดรเจน) จนถึง 118 (ออกาเนสซอน) ได้รับการค้นพบหรือสังเคราะห์ขึ้นมาแล้ว ทำให้ตารางธาตุในปัจจุบันมีครบทั้ง 7 คาบ ธาตุ 94 ตัวแรกพบได้ในธรรมชาติ'
toks = deepcut.tokenize(text)
print(toks)
```

```
# see https://thainlp.org/pythainlp/docs/2.0/index.html

from pythainlp.tokenize import word_tokenize

text = "โอเคบ่พวกเรารักภาษาบ้านเกิด"
words = word_tokenize(text, engine="newmm")
print(words)
```
