```python
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
X = np.random.randn(200, 2)
y = X[:, 1] > np.absolute(X[:, 0])
y = np.where(y, -1, 1)
plt.scatter(X[:, 0],X[:, 1])
```

```python
svm_clf = SVC(kernel = 'rbf')
svm_clf.fit(X,y)
plot_decision_regions(X, y, svm_clf, colors = 'blue,magenta')
plt.show()
```

```python
svm_clf = SVC(kernel = 'sigmoid')
svm_clf.fit(X,y)
plot_decision_regions(X, y, svm_clf, colors = 'blue,magenta')
plt.show()
```

```python
svm_clf = SVC(kernel = 'poly')
svm_clf.fit(X,y)
plot_decision_regions(X, y, svm_clf, colors = 'blue,magenta')
plt.show()
```

```python
svm_clf = SVC(kernel = 'linear')
svm_clf.fit(X,y)
plot_decision_regions(X, y, svm_clf, colors = 'blue,magenta')
plt.show()
```
