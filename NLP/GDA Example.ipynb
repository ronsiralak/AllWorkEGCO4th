```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_samples = 500
X0 = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], num_samples)
X1 = np.random.multivariate_normal([6, 3], [[1, 0], [0, 1]], num_samples)
plt.scatter(X0[:,0], X0[:,1], color = 'r')
plt.scatter(X1[:,0], X1[:,1], color = 'b')
plt.show()
```

```python
phi = 0.5
mu0 = np.average(X0, axis = 0)
mu1 = np.average(X1, axis = 0)
mu0, mu1
```

```python
diff0 = X0 - mu0
diff1 = X1 - mu1
A = np.matmul(diff0.reshape(num_samples,2,1),diff0.reshape(num_samples,1,2))
B = np.matmul(diff1.reshape(num_samples,2,1),diff1.reshape(num_samples,1,2))
sigma = np.average(np.append(A,B, axis = 0),axis = 0)
sigma
```
