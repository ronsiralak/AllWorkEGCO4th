```
import nltk
```

```
sent = "the quick brown fox jumps over the lazy dog the quick brown cat ran away the quick car drive fast"
words = sent.split(" ")
```

```
print(words)
```

```
word_trigrams = nltk.trigrams(words)
```

```
condition_pairs = [ ( (trigram[0], trigram[1]), trigram[2] ) for trigram in word_trigrams ] # ((w1,w2),w3)
```

```
print(condition_pairs) # p(w3|w2,w1)
```

```
cfd = nltk.ConditionalFreqDist(condition_pairs)
```

```
cfd.conditions() # all the w1, w2 in p(w3|w2,w1)
```

```
# syntax is cfd[condition][next word]
cfd[('the', 'quick')]['brown'] # count the occurance of "the quick brown". We have "the quick brown fox" and "the quick brown cat"
```

```
cfd[('the', 'quick')]
```

```
cfd[('the', 'quick')].values()
```

```
sum(cfd[('the', 'quick')].values()) # count the occurance of "the quick". We have "the quick brown fox", "the quick brown cat" and "the quick car"
```

```
# w1 = the, w2 = quick, w3 = brown
# p(w3|w2,w1) = p(brown|quick,the)
prob = cfd[('the', 'quick')]['brown'] / sum(cfd[('the', 'quick')].values())
print(prob)
```

```
cfd[('dog', 'lazy')] # the sequence "dog lazy" never occurs
```

```
sum( cfd[('dog', 'lazy')].values() ) # the sequence that never occurs havs count of 0
```
