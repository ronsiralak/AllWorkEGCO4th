```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(200, 2, centers = 2, cluster_std = 1.5)
```

```python
colors = np.array(['r','b'])
plt.scatter(X[:,0], X[:, 1], c = colors[y])
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,y)
```

```python
model.sigma_
```

```python
cov = np.array([[4,-3.6],
                [-3.6,4]])
mu1  = np.array([0,0])
mu2  = np.array([0,3])

n = 500

x1 = np.random.multivariate_normal(mu1, cov, n).T
x2 = np.random.multivariate_normal(mu2, cov, n).T
X = np.vstack([x1.T, x2.T])
y = np.hstack([[0]*n, [1]*n])
```

```python
plt.scatter(X[:n, 0],X[:n, 1], color = 'r')
plt.scatter(X[n:, 0],X[n:, 1], color = 'b')
plt.plot()
```

```python
model = GaussianNB()
model.fit(X,y)
```

```python
model.sigma_
```

```python
y_fitted = model.predict(X)
colors = np.array(['r','b'])
plt.scatter(X[:, 0],X[:, 1], color = colors[y_fitted])
plt.plot()
```

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.naive_bayes import MultinomialNB
```

```python
data = fetch_20newsgroups()
data.target_names
```

```python
categories = ['comp.graphics', 'rec.autos','sci.crypt']
train = fetch_20newsgroups(subset = 'train', categories = categories)
test = fetch_20newsgroups(subset = 'test', categories = categories)
```

```python
len(train.data), len(train.target), len(test.data), len(test.target)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
```

```python
model = make_pipeline(CountVectorizer(), MultinomialNB())
model.fit(train.data, train.target)
test_result = model.predict(test.data)
print(accuracy_score(test_result, test.target))
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(train.data, train.target)
test_result = model.predict(test.data)
print(accuracy_score(test_result, test.target))
```
