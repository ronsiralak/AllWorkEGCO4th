```python
import pandas as pd
import math
```

```python
def tf_idf(str1,str2):
  list1 = list(str1.split(" "))
  list2 = list(str2.split(" "))
  merged_list = list(set(list1+list2))
  #merged_list.sort()
  tf_d1 = tf_d_fun(merged_list,list1)
  #print(tf_d1)
  tf_d2 = tf_d_fun(merged_list,list2)
  #print(tf_d2)
  nt = nt_fun(tf_d1,tf_d2)
  #print(nt)
  idf = idf_fun(nt)
  #print(idf)
  tf_idf1 = tf_idf_fun(idf,tf_d1)
  #print(tf_idf1)
  tf_idf2 = tf_idf_fun(idf,tf_d2)
  #print(tf_idf2)
  df = showtable(merged_list,tf_d1,tf_d2, nt, idf,tf_idf1,tf_idf2)
  return df
```

```python
def tf_d_fun(merged_list,listn):
  tf_d = [0]*len(merged_list)
  for i in range (0,len(merged_list)):
    count = 0
    for j in range (0,len(listn)):
      if(merged_list[i] == listn[j]):
        count = count + 1
    tf_d[i] = count
  return tf_d
```

```python
def nt_fun(tf_d1,tf_d2):
  nt = [0]*len(tf_d1)
  for i in range (0,len(tf_d1)):
    if(tf_d1[i]>0):
      nt[i] = nt[i] + 1
    if (tf_d2[i]>0):
      nt[i] = nt[i] + 1
  return nt
```

```python
def idf_fun(nt):
  idf = [0]*len(nt)
  for i in range (0,len(nt)):
    idf[i] = math.log(2/nt[i],10)
  return idf
```

```python
def tf_idf_fun(idf,tf_d):
  tf_idf = [0]*len(idf)
  for i in range (0,len(idf)):
    tf_idf[i] = idf[i]*tf_d[i]
  return tf_idf
```

```python
def showtable(merged_list,tf_d1,tf_d2, nt, idf,tf_idf1,tf_idf2):
  data = {'term':merged_list, 
        'tf(t,d1)':tf_d1, 
        'tf(t,d2)':tf_d2, 
        'nt':nt,
        'idf':idf,
        'tf-idf(d1)':tf_idf1,
        'tf-idf(d2)':tf_idf2}
 
  # Create DataFrame
  df = pd.DataFrame(data)
  return df
```
